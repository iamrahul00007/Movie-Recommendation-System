## Import The Required Libraries ##

```python
import pandas as pd
import numpy as np
```

## Load the Files ##

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies.head()
```

```python
print("Credits:", credits.shape)
print("Movies Dataframe:", movies.shape)
```

## Now the Operations begins.... ##

```python
credits_column_renamed = credits.rename(index = str, columns={"movie_id":"id"})
movies_merge = movies.merge(credits_column_renamed, on = 'id')
print(movies_merge.head())
```

```python
movies_cleaned_df = movies_merge.drop(columns = ['homepage','title_x','title_y','status','production_countries'])
print(movies_cleaned_df.head())
print(movies_cleaned_df.info())
print(movies_cleaned_df.head(1)['overview'])
```

```python
movies_cleaned_df['overview'] = movies_cleaned_df['overview'].fillna('')
```

## Import the TF-IDF library for requirements ##

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfv = TfidfVectorizer(min_df=3, max_features=None, strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}', ngram_range=(1, 3), stop_words = 'english')
```

```python
tfv_matrix = tfv.fit_transform(movies_cleaned_df['overview'])
print(tfv_matrix)
print(tfv_matrix.shape)
```

## Requiring a activation function for mimicing the neural networks ##

```python
from sklearn.metrics.pairwise import sigmoid_kernel
```

```python
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
print(sig[0])
```

```python
# Reverse mapping of indices and movie titles
indices = pd.Series(movies_cleaned_df.index, index=movies_cleaned_df['original_title']).drop_duplicates()
print(indices)
print(indices['Avatar'])
print(sig[0])
print(list(enumerate(sig[indices['Avatar']])))
print(sorted(list(enumerate(sig[indices['Avatar']])), key=lambda x: x[1], reverse=True))
```

```python
def give_recomendations(title, sig=sig):
    # Getting the index along to original_title
    idx = indices[title]
    sig_scores = list(enumerate(sig[idx]))
    # Sort the movies
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    # To Show first 10 recomended movies...
    sig_scores = sig_scores[1:11]
    movie_indices = [i[0] for i in sig_scores]
    # Top 10 most similar movies
    return movies_cleaned_df['original_title'].iloc[movie_indices]
```

```python
print(give_recomendations("Tears of the Sun"))
```
